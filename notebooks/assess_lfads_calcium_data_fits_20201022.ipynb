```python
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
import os.path as path
```

```python
loss_file = "C:\\tmp\\lorenz_100\\lfads\\cenc0_cont0_fact3_genc64_gene64_glat64_ulat0_orion-\\loss.pkl"
with open(loss_file,'rb') as f:
    loss_dict = pkl.load(f)
print(loss_dict.keys())
```

```python
loss_dict['valid'].keys()
```

```python
def plot_lfads_loss(loss_dict):
    f, ax = plt.subplots(1,1,dpi=100)
    # training loss
    plt.plot(loss_dict['train']['recon'],label='train: recon.',color='b',linestyle=':')
    plt.plot(loss_dict['train']['kl'],label='train: KL',color='b',linestyle='-.')
    plt.plot(loss_dict['train']['total'],label='train: total',color='b',linestyle='-')
    # validation loss
    plt.plot(loss_dict['valid']['recon'],label='validation: recon.',color='r',linestyle=':')
    plt.plot(loss_dict['valid']['kl'],label='validation: KL',color='r',linestyle='-.')
    plt.plot(loss_dict['valid']['total'],label='validation: total',color='r',linestyle='-')
    ax.set_yscale('log')
    ax.legend(loc=0)
    # labels
    ax.set_xlabel('Training iter.')
    ax.set_ylabel('loss')
    ax.set_title('LFADS Training Loss: Calcium Imaging')
```

```python
plot_lfads_loss(loss_dict)
```
